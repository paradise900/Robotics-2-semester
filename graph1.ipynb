```python
from math import pi
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.polynomial import Polynomial

napr = 50
param = []
for i in range(10):
    param.append([])
f = open(f'res2/data{napr}')
a = [x for x in f.readlines()]
f.close()
data = []
for i in range(len(a)):
    data.append(a[i][:-1].split(','))
n = len(data)
```

```python
for i in range(len(data)):
    data[i][1] = int(data[i][1]) * pi / 180
    data[i][0] = float(data[i][0]) 
    data[i][2] = int(data[i][2]) * pi / 180

data = pd.DataFrame(data)
```

```python
def plot_problem(ax, x_data, y_data):
    ax.plot(x_data, y_data, linewidth=0.5, label=f"Исходная зависимость", color='red')
    ax.set_xlabel("$time, s$")
    ax.set_ylabel("$angel speed, rad/s$")
    ax.set_title(f"U = {napr}")
    ax.legend()

xh = np.linspace(0, 1, 1000) 

fig, ax = plt.subplots(layout="tight")
approximation = Polynomial.fit(data[0], data[2], 5)
plot_problem(ax, data[0], data[2])
ax.plot(xh, approximation(xh), color="green", label="Аппроксимация", linewidth=0.7)
ax.legend()
# plt.savefig(f'res2/{napr}_2.svg')
```
