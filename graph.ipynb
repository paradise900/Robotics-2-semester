```python
from math import pi
import numpy as np
import pandas as pd
napr = 20

param = []
for i in range(10):
    param.append([])
f = open(f'Telegram Archive/data{napr}')
a = [x for x in f.readlines()]
f.close()
data = []
for i in range(len(a)):
    data.append(a[i][:-1].split(','))
n = len(data)
data
```

```python

for i in range(len(data)):
    data[i][1] = int(data[i][1]) * pi / 180
    data[i][0] = float(data[i][0]) / 1000
    data[i][2] = int(data[i][2]) 

data = pd.DataFrame(data)
data
```

```python
import matplotlib.pyplot as plt


def plot_problem(ax, x_data, y_data):
    ax.plot(data[0], data[1], linewidth=0.5, label="Исходная зависимость")
    ax.set_xlabel("$x$")
    ax.set_ylabel("$y$")
    


fig, ax = plt.subplots(layout="tight")

plot_problem(ax, data[0], data[1])
ax.set_title(f"Исходные данные")
ax.legend()
```
