```python
import ev3dev2.motor as motor
import time
import math
```

```python
def chek_vol(vol):
    if vol > 100:
        vol = 100
    elif vol < -100:
        vol = -100
    return vol
```

```python
def p_reg(k_p):
    motorA = motor.LargeMotor('outA')
    h = 0.001
    theta_goal = 360
    vol = 100
    timeStart = time.time()
    startPos = motorA.position
    file = open('p_reg', 'w')
    while True:
        timeNow = time.time() - timeStart
        motorA.run_direct(duty_cycle_sp = vol)
        pos = motorA.position - startPos
        file.write(str(timeNow) + ',' + str(pos))
        timeNow1 = time.time() - timeStart
        if timeNow1 - timeNow < h:
            time.sleep(h - timeNow1 + timeNow)
        if time.time() - timeStart > 10:
            motorA.run_direct(duty_cycle_sp = 0)
            break
        vol = k_p * (theta_goal - pos)
        vol = chek_vol(vol)
    file.close()
```

```python
def pi_reg(k_p, k_i):
    motorA = motor.LargeMotor('outA')
    h = 0.001
    theta_goal = 360
    vol = 100
    timeStart = time.time()
    startPos = motorA.position
    integral = 0
    pos = []
    file = open('pi_reg', 'w')
    
    motorA.run_direct(duty_cycle_sp = vol)
    timeNow = time.time() - timeStart
    if timeNow < h:
        time.sleep(h - timeNow)
        timeNow = time.time() - timeStart
    pos.append(motorA.position - startPos)
    file.write(str(timeNow) + ',' + str(pos[-1]))
    timeLast = timeNow

    while True:
        motorA.run_direct(duty_cycle_sp = vol)
        timeNow = time.time() - timeStart
        if timeNow - timeLast < h:
            time.sleep(h - timeNow + timeLast)
            timeNow = time.time() - timeStart
    
        pos[-2] = pos[-1]
        pos[-1] = motorA.position - startPos
        file.write(str(timeNow) + ',' + str(pos[-1]))
       
        if time.time() - timeStart > 10:
            motorA.run_direct(duty_cycle_sp = 0)
            break

        integral += (pos[-1] + pos[-2]) / (timeNow - timeLast)
        vol = k_p * (theta_goal - pos) + k_i * integral
        vol = chek_vol(vol)
        timeLast = timeNow   
    file.close()
```
