```python
import matplotlib.pyplot as plt
```

```python
num = '_perelet'
```

```python
f = open(f'res{num}/p_reg')
t = []
ang = []

for x in f:
    a, b = x.split(',')
    t.append(float(a))
    ang.append(int(b))

def plot_problem(ax, x_data, y_data):
    ax.plot(x_data, y_data, linewidth=0.5, label=f"Исходная зависимость", color='red')
    ax.axline((0, 360), (3, 360), linestyle='--')
    ax.legend()



fig, ax = plt.subplots(layout="tight")

plot_problem(ax, t, ang)
ax.legend()
```
