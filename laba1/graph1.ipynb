```python
from math import pi
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.polynomial import Polynomial

napr = 60
param = []
for i in range(10):
    param.append([])
f = open(f'newres 2/data{napr}')
a = [x for x in f.readlines()]
f.close()
data = []
for i in range(len(a)):
    data.append(a[i][:-1].split(','))
n = len(data)
```

```python
for i in range(len(data)):
    data[i][1] = int(data[i][1]) * pi / 180
    data[i][0] = float(data[i][0]) 
    data[i][2] = int(data[i][2]) * pi / 180

data = pd.DataFrame(data)
```

```python
def plot_problem(ax, x_data, y_data):
    ax.plot(x_data, y_data, linewidth=0.5, label=f"Исходная зависимость", color='red')
    ax.set_xlabel("$time, s$")
    ax.set_ylabel("$angel, rad$")
    ax.set_title(f"U = {napr}")
    ax.legend()

xh = np.linspace(0, 1, 1000) 

fig, ax = plt.subplots(layout="tight")
approximation = Polynomial.fit(data[0], data[1], 1)
plot_problem(ax, data[0], data[1])
ax.plot(xh, approximation(xh), color="green", label="Аппроксимация", linewidth=0.7)
ax.legend()
# plt.savefig(f'new_graphics/{napr}_1.svg')
```

```python
def plot_problem(ax, x_data, y_data):
    ax.plot(x_data, y_data, linewidth=0.5, label=f"Исходная зависимость", color='blue')
    ax.set_xlabel("$time, s$")
    ax.set_ylabel("$angel speed, rad/s$")
    ax.set_title(f"U = {napr}")
    ax.legend()

xh = np.linspace(0, 1, 1000) 

fig, ax = plt.subplots(layout="tight")
approximation = Polynomial.fit(data[0], data[2], 10)
plot_problem(ax, data[0], data[2])
ax.plot(xh, approximation(xh), color="red", label="Аппроксимация", linewidth=0.7)
ax.legend()
plt.rc('font', size=14) #controls default text size
plt.rc('axes', titlesize=14) #fontsize of the title
plt.rc('axes', labelsize=14) #fontsize of the x and y labels
plt.rc('xtick', labelsize=14) #fontsize of the x tick labels
plt.rc('ytick', labelsize=14) #fontsize of the y tick labels
plt.rc('legend', fontsize=14) 
plt.savefig(f'{napr}_2.svg')
```
