```python
from math import pi
import numpy as np
import pandas as pd
napr = 20

param = []
for i in range(10):
    param.append([])
f = open(f'Telegram Archive/data{napr}')
a = [x for x in f.readlines()]
f.close()
data = []
for i in range(len(a)):
    data.append(a[i][:-1].split(','))
n = len(data)
```

```python

for i in range(len(data)):
    data[i][1] = int(data[i][1]) * pi / 180
    data[i][0] = float(data[i][0]) / 1000
    data[i][2] = int(data[i][2]) * pi / 180

data = pd.DataFrame(data)
```

```python
from numpy.polynomial import Polynomial

def plot_problem(ax, x_data, y_data):
    ax.plot(x_data, y_data, linewidth=0.5, label="Исходная зависимость")
    ax.set_xlabel("$x$")
    ax.set_ylabel("$y$")
    ax.set_title(f"Исходные данные")
    ax.legend()

xh = np.linspace(0, 0.001, 100) 

fig, ax = plt.subplots(layout="tight")
approximation = Polynomial.fit(data[0], data[1], 1)
plot_problem(ax, data[0], data[1])
ax.plot(xh, approximation(xh), color="green", label="Аппроксимация")
ax.set_title(f"$n = $")
ax.legend()
```
