```python
import matplotlib.pyplot as plt
```

```python
for i in range(5):
    f = open(f'res2/coord{i}')
    x = []
    y = []
    arr = f.readlines()
    for i in arr:
        a, b, c = i.split(',')
        x.append(float(b) / 2)
        y.append(float(c) / 2)

    plt.plot(x, y)
    f.close()
    
    
plt.scatter([0, 0, 0.5, -0.5, 0], [0.5, -0.5, 0, 0, 0])
```
